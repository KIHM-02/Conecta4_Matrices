```python
def define_matriz(matriz):
    print("Para poder avanzar, necesitamos conocer el tablero")
    filas = int(input('Ingresa la cantidad de filas:\n--> '))
    columnas = int(input('Ingresa la cantidad de columnas:\n--> '))
    
    print("\n\nCreando tablero....................\n")
    
    for iterations in range(filas):
        lista = ['*'] * columnas
        matriz.append(lista)
        
    __display_game__(filas, matriz)
    
    print("\nTablero creado.......\n\n")
    return matriz
    
def __display_game__(filas, matriz):
    print()
    for iterations in range(filas):
        print(matriz[iterations])
    
def __define_player__():
    player_one = str(input("Primer jugador, escoja un caracter del teclado\n-->"))
    player_two = str(input("Segundo jugador, escoja un caracter del teclado\n-->"))
    
    print(f'Primer jugador: {player_one}\nSegundo jugador: {player_two}')
    return player_one, player_two
    
def place_token(player, matriz):
    
    __display_game__(len(matriz), matriz)
    
    fila = int(input('\n\nIngresa la fila a colocar la ficha:\n-> '))
    columna = int(input('Ingresa la columna a colocar la ficha:\n-> '))
    
    while matriz[fila][columna] != '*':
        print("La casilla ya esta ocupada, vuelve a ingresar otra ubicacion")
        fila = int(input('Ingresa la fila a colocar la ficha:\n-> '))
        columna = int(input('Ingresa la columna a colocar la ficha:\n-> '))
        
    matriz[fila][columna] = player
    
def check_winner(matriz):
    for analysis in (horizontal_analisys, vertical_analisys, analysis_diagonal, analysis_diagonal_inverse):
        resultado = analysis(matriz)
        if resultado[0]:  # Si hay un ganador, retorna de inmediato
            return resultado
    return (False, '*')  # No hay ganador

    
def horizontal_analisys(matriz):
    fila = 0
    ganador = False
    ficha = '*'
    
    while fila < len(matriz) and ganador != True:
        dato = 0
        contador = 0
        
        while dato < (len(matriz[fila])-1) and contador < 3:
            if matriz[fila][dato] != '*':
                if matriz[fila][dato] == matriz[fila][dato+1]:
                    contador += 1
                else:
                    contador = 0
                    
                if contador == 3:
                    ganador = True
                    ficha = matriz[fila][dato]
                    
            dato += 1
        fila += 1
        
    return (ganador, ficha)
    
def vertical_analisys(matriz):
    matriz_volteada = [list(columna) for columna in zip(*matriz)]
    resultado = horizontal_analisys(matriz_volteada)
    return resultado


def analysis_diagonal(matriz):
    filas = len(matriz)
    columnas = len(matriz[0]) if filas > 0 else 0

    # Evaluar diagonales ↘ (abajo-derecha)
    for inicio_fila in range(filas - 3):
        fila_diagonal, columna = inicio_fila, 0
        contador = 0

        while fila_diagonal < filas and columna < columnas:  # Mientras esté en los límites
            if matriz[fila_diagonal][columna] != '*' and fila_diagonal > inicio_fila and matriz[fila_diagonal][columna] == matriz[fila_diagonal - 1][columna - 1]:
                contador += 1
            else:
                contador = 0

            if contador == 3:
                return (True, matriz[fila_diagonal][columna])

            fila_diagonal += 1
            columna += 1

    # Evaluar diagonales ↖ (arriba-izquierda)
    for inicio_fila in range(filas - 1, 2, -1):
        fila_diagonal, columna = inicio_fila, columnas - 1
        contador = 0

        while fila_diagonal >= 0 and columna >= 0:  # Mientras esté en los límites
            if matriz[fila_diagonal][columna] != '*' and fila_diagonal < inicio_fila and matriz[fila_diagonal][columna] == matriz[fila_diagonal + 1][columna + 1]:
                contador += 1
            else:
                contador = 0

            if contador == 3:
                return (True, matriz[fila_diagonal][columna])

            fila_diagonal -= 1
            columna -= 1

    return (False, '*')

	
def bottom_inverse(matriz):
    fila = 0
    fila_diagonal = 0
    contador = 0
    # Evaluación de diagonales ↙ (abajo-izquierda)
    for fila in range(len(matriz) - 3):  # No tiene caso evaluar los últimos 3 casos porque no se juntan las 4 fichas
        print("Soy la fila", fila)
        columna = 0
        fila_diagonal = fila

        for columna in range((len(matriz[fila]) - 1), -1, -1):
            print("Soy la columna", columna)
            if fila_diagonal < (len(matriz) - 1) and columna > 0:  # Verifica límites antes de acceder
                if matriz[fila_diagonal][columna] != '*' and matriz[fila_diagonal][columna] == matriz[fila_diagonal + 1][columna - 1]:
                    print(f'Fila diag: {fila_diagonal} Columna: {columna}')
                    contador += 1
                else:
                    contador = 0
                    print(f'Ya evalué en Fila diag: {fila_diagonal} Columna: {columna}')
                
                if contador == 3:
                    return (True, matriz[fila_diagonal][columna])

            fila_diagonal += 1
            print(f'La diagonal aumentó en uno {fila_diagonal}')
    
    return (False, '*')

    
def top_inverse(matriz):
    # Evaluación de diagonales ↖ (arriba-derecha)
    for fila in range((len(matriz) - 1), -1, -1):  # Se evalúa desde el último valor de la matriz para ir subiendo hasta llegar a la fila 0
        print(fila)
        columna = 0
        fila_diagonal = fila

        for columna in range(len(matriz[fila]) - 1):  # Se evalúa desde la última columna para ir subiendo hasta llegar al 0
            if fila_diagonal != 0:  # Es el último valor a evaluar, si este pasara de 0, tendríamos un desbordamiento de la matriz
                if matriz[fila_diagonal][columna] != '*' and matriz[fila_diagonal][columna] == matriz[fila_diagonal - 1][columna + 1]:
                    print(f'Fila diag: {fila_diagonal} Columna: {columna}')
                    contador += 1
                else:
                    contador = 0
                    print(f'Ya evalué en Fila diag: {fila_diagonal} Columna: {columna}')
                
                if contador == 3:
                    return (True, matriz[fila_diagonal][columna])

            fila_diagonal -= 1
            print(f'La diagonal aumentó en uno {fila_diagonal}')

    return (False, '*')
    
################### TODO: Fit all the diagonal analisys in the game function
def analysis_diagonal_inverse(matriz):
    winner = bottom_inverse(matriz)
    
    if winner[0] is True:
        print("\n\n\n--------", winner)
        return winner
    else:
        winner = top_inverse(matriz)
        print("\n\n\n--------", winner)
        if winner[0] is True:
            return winner
    
    return (False, '*')
    
def game(player_one, player_two, matriz):
    ganador = False
    turno = False
    
    while ganador == False:
        if turno == False:
            print(f'\nTurno del primero jugador: {player_one}')
            place_token(player_one, matriz)
            turno = True
        else:
            print(f'\nTurno del segundo jugador: {player_two}')
            place_token(player_two, matriz)
            turno = False
        
        resultado = check_winner(matriz)
        
        if resultado[0] == True:
            ganador = True
            print(f'\n---->El ganador es: {resultado[1]}')
        
    __display_game__(len(matriz), matriz)
        
        
def __display_instructions__():
    instructions = """\nInstrucciones......\nEl juego 4 en raya, también conocido como Conecta 4, se juega con
fichas de diferentes caracteres (1 por jugador) en una casilla. El objetivo es alinear cuatro
fichas en línea, ya sea horizontal, vertical o diagonalmente.

Para agregar la casilla, debes registrar la fila y la columna en la posicion que se desea colocar"""
    
    print(instructions)
    
def __read_option__(menu):
    while (clave := int(input("Opcion:\n-->"))) not in menu:
        print("Opcion incorrecta, vuelva a intentarlo")
    return clave

def __exit__():
    print("\nAdios")

def display_menu(menu, opcion_start, opcion_salida):
    opcion = None
    while opcion != opcion_salida:
        print("Ingresa la opcion deseada: ")
        
        for clave in sorted(menu):
            print(f'{clave} {menu[clave][0]}')
        
        opcion = __read_option__(menu)
        
        if opcion == opcion_start:
            print() #Salto de linea
            return True
        else:
            menu[opcion][1]()
            print() #Salto de linea
    
    return False
        
def main():
    opcion = None
    opcion_salida = 3
    opcion_start = 2
    player_one = ''
    player_two = ''
    menu = {
        1: ('Instrucciones', __display_instructions__),
        2: ('Iniciar partida', ),
        3: ('Salir', __exit__)
    }
    matriz = []

    
    print("Hola, bienvenidos al juego 4 en raya")
    
    if len(matriz) == 0:
        matriz = define_matriz(matriz)
    
    start = display_menu(menu, opcion_start, opcion_salida)
    
    if start != True:
        return #exit from the program
    
    player_one, player_two = __define_player__()
    game(player_one, player_two, matriz)
    
        
```
